```python
pk_2['Name'] = pk_2['Name'].str.replace(".*(?=Mega)", "")
```

```python
pk_2.head()
```

Pengubahan dikolom 'Name' agar tidak terjadi pengulangan nama seperti "BeedrillMega Beedrill" dengan perintah ganti kata didepan 'Mega' menjadi'' dengan perintah pk_2['Name'] = pk_2['Name'].str.replace(".*(?=Mega)", "")

```python
pk_2['Type 2'].fillna(pk_2['Type 1'], inplace=True)
```

Perintah untuk mengisi kolom 'Type 2' dengan 'Type 1' apa bila bernilai NaN

```python
pk_2.head()
```

```python
pk_3=pk_2.set_index('Name')
```

```python
pk_4=pk_3.drop('Generation',axis=1)
pk_4.head()
```
